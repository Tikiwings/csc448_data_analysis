```python
import numpy as np
import pandas as pd
import sys
import json

def get_data(data_file):
#     print("analyzing given file...")
    
    states = []
    recovered = []
    deaths = []
    hospitalized = []
    totalTestResults = []
    total = []


    try:
        f = open(data_file)
        data = json.load(f)
    except:
        print("Error in retrieving file")
        exit()


    for n in data:
        params = (n['state'],n['recovered'],n['death'],n['hospitalized'],n['totalTestResults'],n['total'])
        states.append(0 if params[0] == None else params[0])
        recovered.append(0 if params[1] == None else params[1])
        deaths.append(0 if params[2] == None else params[2])
        hospitalized.append(0 if params[3] == None else params[3])
        totalTestResults.append(0 if params[4] == None else params[4])
        total.append(0 if params[5] == None else params[5])
    f.close()
    
#     print(states)
#     print(recovered)
#     print(deaths)
#     print(hospitalized)
#     print(totalTestResults)
#     print(total)

    df = pd.DataFrame({'recovered':recovered,
                       'deaths':deaths,
                       'hospitalized':hospitalized,
                       'totalTestResults':totalTestResults,
                       'total':total})
    df.index = states
    return df
    
```
